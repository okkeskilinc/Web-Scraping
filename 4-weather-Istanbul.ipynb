```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
def transform(data):
    transformed_data=(data-32)/1.8
    return transformed_data
```

```python
url="https://weather.com/weather/tenday/l/33d1e415eb66f3e1ab35c3add45fccf4512715d329edbd91c806a6957e123b49"
html_data=requests.get(url)
html_data
```

```python
soup=BeautifulSoup(html_data.content,"html.parser")
```

```python
ten_day=soup.find(class_="DailyForecast--DisclosureList--msYIJ")
forecast_items=ten_day.find_all(class_="Disclosure--themeList--25Q0H")
to_day=forecast_items[0]
print(to_day.prettify())
```

```python
period=to_day.find(class_="DailyContent--daypartName--1bzYn").text
temp=to_day.find(class_="DailyContent--temp--3d4dn").text
desc_short=to_day.find(class_="DailyContent--narrative--hplRl").text

print(period)
print(temp)
print(desc_short)
```

```python
period_tags=ten_day.select(".Disclosure--themeList--25Q0H .DailyContent--daypartName--1bzYn")
periods=[pt.text for pt in period_tags]

temp_tags=ten_day.select(".Disclosure--themeList--25Q0H .DailyContent--temp--3d4dn")
temps=[tm.text for tm in temp_tags]

desc_tags=ten_day.select(".Disclosure--themeList--25Q0H .DailyContent--narrative--hplRl")
descs=[ds.text for ds in desc_tags]
```

```python
weather=pd.DataFrame({
    "period":periods,
    "temp-F":temps,
    "desc":descs
})
weather
```

```python
temp_nums=weather["temp-F"].str.extract("(\d+)", expand=False)
```

```python
weather["Celcius"]=temp_nums.astype('int')
weather["Celcius"]=transform(weather["Celcius"])
```
